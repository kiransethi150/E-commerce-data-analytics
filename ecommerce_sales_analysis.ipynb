## Data Summary & Cleaning

```python
df = pd.read_csv("ecommerce_dataset.csv")
df.head()
```

## KPIs

```python
# Total Revenue
total_revenue = df['TotalAmount'].sum()
print(f'Total Revenue: ${total_revenue:,.2f}')

# Total Orders
print(f"Total Orders: {df['OrderID'].nunique()}'")

# Average Order Value
print(f"Average Order Value: ${df['TotalAmount'].mean():.2f}")
```

## Sales Over Time

```python
#convert 'OrderDate' column to datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

monthly_sales = df.resample('M', on='OrderDate')['TotalAmount'].sum().reset_index()
px.line(monthly_sales, x='OrderDate', y='TotalAmount', title='Monthly Revenue Trend')
```

## Top Selling Products

```python
top_products = df.groupby('ProductName')['TotalAmount'].sum().sort_values(ascending=False).head(10).reset_index()
px.bar(top_products, x='ProductName', y='TotalAmount', title='Top 10 Products by Revenue')
```

## Revenue by Region

```python
region_sales = df.groupby('Region')['TotalAmount'].sum().reset_index()
px.pie(region_sales, names='Region', values='TotalAmount', title='Revenue Share by Region')
```
